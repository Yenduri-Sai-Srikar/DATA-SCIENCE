reading in chunks(skip+nrows)

```python
list_of_dataframes = []
rows_in_a_chunk = 10
num_chunks =5
df_dummy = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DATA SCIENCE/csv/Boston_housing.csv",nrows=2)
col_names = df_dummy.columns
for i in range (0,num_chunks*rows_in_a_chunk):
  df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DATA SCIENCE/csv/Boston_housing.csv",header=0,skiprows=2)
  list_of_dataframes.append(df)
```
